```python
import os
import re
import aocd
import string
import numpy as np
import pandas as pd
from copy import deepcopy
from aocd import submit
# My solution

id_sum = 0
rooms = []
alph = [x for x in string.ascii_lowercase]
room_names = []

data = aocd.get_data(day=4, year=2016).split('\n')
#data = ['qzmt-zixmtkozy-ivhz-343[abdce]']
for line in data:

    letters = line[:-1].split('-')
    sector_id, checksum = letters[-1].split('[')
    sector_id = int(sector_id)
    
    room_name = []
    for word in letters[:-1]:
        rolled_alph = np.roll(np.array(alph), -sector_id)
        room_name += [''.join(rolled_alph[[alph.index(x) for x in word]])]
    room_name = ' '.join(room_name)

    
    
    joined = ''.join(letters[:-1])
    df = pd.DataFrame([x for x in joined], columns=['char'])
    df['count'] = 1
    unique = df.groupby('char')['count'].count()
    unique = unique.reset_index().sort_values(['count','char'],ascending=[False,True])
    most_common = ''.join(unique['char'].values[:5])

    if most_common == checksum:
        id_sum += sector_id
        room_names += [[room_name, sector_id]]
        print(room_name, sector_id)

id_sum
```

```python
from aocd import data, submit
from aocd.get import get_data

#Part 1
data = get_data(day=4, year=2020).split('\n')
```

```python
passport = ''


answer = 0
for line in data:
    passport += line + ' '
    if line == '':
        required = ['byr','iyr','eyr','hgt','hcl','ecl','pid']
        optional = ['cid']
        passport = passport.split(' ')
        keys = [x.split(':')[0] for x in passport if x != '']
        
        invalid_key = False
        for key in keys:
            if key in required:
                required.remove(key)
            elif key in optional:
                optional.remove(key)
            else:
                invalid_key = True

        if (not invalid_key) and (len(required) == 0):
            answer += 1
            
        print(passport, keys, answer)
        passport = ''
        
        
```

```python
import re

#Part 2
passport = ''

re_patterns = {
    'byr':'^\d{4}$',
    'iyr':'^\d{4}$',
    'eyr':'^\d{4}$',
    'hgt':'^\d{2,3}(cm|in)$',
    'hcl':'^#[0-9,a-f]{6}$',
    'ecl':'^(amb|blu|brn|gry|grn|hzl|oth)$',
    'pid':'^\d{9}$'
}

answer = 0
for line in data:
    passport += line + ' '
    if (line == '') | (line == data[-1]):
        required = ['byr','iyr','eyr','hgt','hcl','ecl','pid']
        optional = ['cid']
        passport = passport.split(' ')
        keys = [x.split(':')[0] for x in passport if x != '']
        values = [x.split(':')[1].strip() for x in passport if x != '']
        
        invalid_key = False
        for key,value in zip(keys,values):
            if key in required:
                required.remove(key)
            elif key in optional:
                optional.remove(key)
            else:
                invalid_key = True
                continue
                
            if key == 'cid':
                continue
                
            if re.match(re_patterns[key],value) is None:
                invalid_key = True
                continue
            
            if key == 'byr':
                if (int(value) > 2002) | (int(value) < 1920):
                    invalid_key=True
            if key == 'iyr':
                if (int(value) > 2020) | (int(value) < 2010):
                    invalid_key=True
            if key == 'eyr':
                if (int(value) > 2030) | (int(value) < 2020):
                    invalid_key=True
            if key == 'hgt':
                if 'cm' in value:
                    min_, max_ = 150,193
                else:
                    min_, max_ = 59,76
                    
                height = int(re.findall('\d+',value)[0])
                if (height > max_) | (height < min_):
                    invalid_key = True
            
            
                    
            

        if (not invalid_key) and (len(required) == 0):
            answer += 1
            
        print(passport, keys, answer)
        passport = ''
```

```python
from aocd import data, submit
from aocd.get import get_data
import numpy as np
import pandas as pd
#Part 1
data = get_data(day=7, year=2020).split('\n')
color = [x.split(',') for x in data]
color = [x.split('contain')[0].replace('bags','').replace('bag','').strip() for x in data if 'contain' in x]
contains = [x.split(',') for x in data]
contains = [[y.split('contain')[1] if 'contain' in y else y for y in x ] for x in contains ]
contains = [[y[2:].replace('bags','').replace('.','').replace('bag','').strip() for y in x] for x in contains]
contains
```

```python
checked = []
to_check_list = ['shiny gold']

while len(to_check_list):
    to_check = to_check_list.pop()
    for col, con in zip(color,contains):
        if to_check in con:
            to_check_list += [col]
    checked += [to_check]
```

```python
import re
contains = [x.split(',') for x in data]
contains = [[y.split('contain')[1] if 'contain' in y else y for y in x ] for x in contains ]
contains = [[y.replace('bags','').replace('.','').replace('bag','').strip() for y in x] for x in contains]
bags = {}

for colo, y in zip(color,contains):
    bags[colo] = {}
    for x in y:
        if 'no other' in x:
            continue
        no = int(x.split(' ')[0])
        col = ' '.join(x.split(' ')[1:])
        bags[colo][col] = no
```

```python
def recursive_(bag_color):
    _ = bags[bag_color]
    return sum(_.values()) + sum([v*recursive_(k) for k,v in _.items()])
recursive_('shiny gold')
```

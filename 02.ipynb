```python
from aocd import data, submit
from aocd.get import get_data
```

```python
import re

no_valid_passwords = 0
#Part 1
data = get_data(day=2, year=2020).split('\n')
for line in data:
    length, letter, password = line.split(' ')
    
    #Some housekeeping
    _min, _max = length.split('-')
    _min, _max = int(_min), int(_max)
    letter = letter[0]
    
    letter_appearances = len(re.findall(letter, password))
    
    if _min <= letter_appearances <= _max:
        no_valid_passwords += 1
        print('Good',line)
    else:
        print('bad', line)
    
```

```python
submit(no_valid_passwords, day=2, year=2020, part='a')
```

```python
# Part 2
no_valid_passwords = 0

for line in data:
    indices, letter, password = line.split(' ')
    
    #Some housekeeping
    idx1, idx2 = indices.split('-')
    idx1 = int(idx1)-1
    idx2 = int(idx2)-1
    letter = letter[0]
    
    matches = 0
    if (len(password)-1 >= idx1):
        if (password[idx1] == letter):
            matches += 1
    if (len(password)-1 >= idx2):
        if (password[idx2] == letter):
            matches += 1
    
    if matches == 1:
        no_valid_passwords += 1
        print('Good',line)
    else:
        print('bad', line)
```

```python
submit(no_valid_passwords, day=2, year=2020, part='b')
```

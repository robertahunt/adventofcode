```python
import os
import re
import aocd
import numpy as np
import pandas as pd
from aocd import submit
from tqdm import tqdm
from collections import Counter


#data = aocd.get_data(day=11, year=2021)#.split('\n')
data = """5665114554
4882665427
6185582113
7762852744
7255621841
8842753123
8225372176
7212865827
7758751157
1828544563""".split('\n')
test = """5483143223
2745854711
5264556173
6141336146
6357385478
4167524645
2176841721
6882881134
4846848554
5283751526""".split('\n')
#data = test
data = np.array([[int(y) for y in x] for x in data])

def get_adjacent(i,j, data):
    corners = []
    if i != 0:
        corners += [(i-1,j)]
    if( i != 0) & (j!=0):
        corners += [(i-1,j-1)]
    if( i != 0) & (j != (len(data[i]) - 1)):
        corners += [(i-1,j+1)]
    if( i != (len(data) - 1)) & (j!=0):
        corners += [(i+1,j-1)]
    if(i != (len(data) - 1)) & (j != (len(data[i]) - 1)):
        corners += [(i+1,j+1)]
    if j != 0:
        corners += [(i,j-1)]
    if i != (len(data) - 1):
        corners += [ (i+1,j) ]
    if j != (len(data[i]) - 1):
        corners += [(i,j+1)]
    return corners

n_steps = 1000

tot_flashed = 0
for step in range(1,n_steps+1):
    flashed = np.zeros(data.shape)
    data += 1
    next_flashed = np.zeros(data.shape)
    next_flashed[data >= 10] = 1
    while (next_flashed != flashed).any():
        _is, _js = np.where((next_flashed == 1)&(flashed == 0))
        for i, j in zip(list(_is),list(_js)):
            adj = get_adjacent(i,j,data)

            for ai, aj in adj:
                data[ai,aj] += 1
        
        flashed = next_flashed
        
        next_flashed = np.zeros(data.shape)
        next_flashed[data >= 10] = 1
        
    tot_flashed += next_flashed.sum()
    data[data >= 10] = 0
    
    if next_flashed.sum() == np.product(next_flashed.shape):
        print('Found when all flash', step)
    
data, flashed,tot_flashed
```

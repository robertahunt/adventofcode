```python
#part 2
data = aocd.get_data(day=5, year=2021).split('\n')
test = """0,9 -> 5,9
8,0 -> 0,8
9,4 -> 3,4
2,2 -> 2,1
7,0 -> 7,4
6,4 -> 2,0
0,9 -> 2,9
3,4 -> 1,4
0,0 -> 8,8
5,5 -> 8,2""".split('\n')
#data = test

data = [x.split('->') for x in data]
data = [y.strip().split(',') for x in data for y in x ]
data = np.array(data).reshape(-1,).reshape(-1,4)
data=pd.DataFrame(data, columns=['x1','y1', 'x2','y2'])
data = data.applymap(int)

_max = data.max().max() + 1
#data = data[(data['x1'] == data['x2']) | (data['y1'] == data['y2'])]

grid = np.zeros((_max, _max))

for name, row in tqdm(data.iterrows(), total=len(data)):
    x1, y1, x2, y2 = row
    x_flag = 1
    if x1 > x2:
        x_flag = -1
    y_flag = 1
    if y1 > y2:
        y_flag = -1
        
    if x1 == x2:
        y1, y2 = min(y2, y1), max(y2, y1)
        for j in range(y1, y2+1):
            grid[j, x1] += 1
    elif y1 == y2:
        x1, x2 = min(x2, x1), max(x2, x1)
        for i in range(x1, x2+1):
            grid[y1, i] += 1
    else:
        for i, j in zip(range(x1, x2+x_flag, x_flag), range(y1, y2+y_flag, y_flag)):
            grid[j, i] += 1

grid, data
len(grid[grid>=2])
```

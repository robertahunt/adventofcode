```python
import os
import re
import aocd
import pandas as pd
from aocd import submit

data = aocd.get_data(day=1, year=2021).split('\n')
data = [int(x) for x in data]
data
```

```python
prev = data[0]
increases = 0
for i in data[1:]:
    if i > prev:
        increases += 1
    prev = i
increases
```

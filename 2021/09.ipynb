```python
def get_corners(i,j, data):
    corners = []
    if i != 0:
        corners += [(i-1,j)]
    if j != 0:
        corners += [(i,j-1)]
    if i != (len(data) - 1):
        corners += [ (i+1,j) ]
    if j != (len(data[i]) - 1):
        corners += [(i,j+1)]
    return corners

data = aocd.get_data(day=9, year=2021).split('\n')
test = """2199943210
3987894921
9856789892
8767896789
9899965678""".split('\n')
#data = test
data = [[int(y) for y in x] for x in data]
data = np.array(data)

lows = []
basins = []
for i in range(len(data)):
    for j in range(len(data[i])):
        corners = []
        if i != 0:
            corners += [(i-1,j)]
        if j != 0:
            corners += [(i,j-1)]
        if i != (len(data) - 1):
            corners += [ (i+1,j) ]
        if j != (len(data[i]) - 1):
            corners += [(i,j+1)]
        
        corner_values = [data[i,j] >= data[ci,cj] for ci,cj in corners]
        _sum = sum(corner_values)
        if _sum == 0:
            #print('found lowpoint', i,j, data[i,j], corner_values)
            lows += [(i,j)]

basins = {x:0 for x in lows}
for i in tqdm(range(len(data))):
    for j in range(len(data[i])):
        #print('i,j',i,j)
        if data[i,j] == 9:
            continue
        _i, _j = i,j
        its = 0
        while (_i,_j) not in lows:
            corners = get_corners(_i,_j,data)
            for ci,cj in corners:
                #print(data[ci,cj], data[_i,_j],data[ci,cj] < data[_i,_j])
                if data[ci,cj] < data[_i,_j]:
                    _i, _j = ci, cj
            its += 1
            if its == 10:
                break
        basins[(_i,_j)] += 1

basins
```

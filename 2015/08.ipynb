```python
n_strs = []
n_codes = []

for line in data:
    line = line[1:-1]
    print(line)
    n_str = 0
    n_code = 2

    line = line.replace('\\','_')
    hexs = re.findall('_x[a-z0-9]{2}', line)
    for h in hexs:
        n_code += 4
        n_str += 1
        print('   ', h)
    line = re.sub('_"', 'å', line)
    line = re.sub('__', 'æ', line)
    line = re.sub('_x[a-f0-9]{2}', '', line)
    print('   ',line)
    for char in line:
        if re.match('[a-z]',char):
            n_code += 1
            n_str += 1
        elif re.match('å',char):
            n_code += 2
            n_str += 1
        elif re.match('æ',char):
            n_code += 2
            n_str += 1
        else:
            print('   a? ',char)
            n_code += 4
            n_str += 1
    print('   ',n_code, n_str)
    n_strs += [n_str]
    n_codes += [n_code]
n_str, n_code
```

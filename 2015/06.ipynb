```python
from aocd import data, submit
from aocd.get import get_data
from tqdm import tqdm
import numpy as np
import pandas as pd
import re
import hashlib
import matplotlib.pyplot as plt
#Part 1
example = """turn on 0,0 through 999,999
toggle 0,0 through 999,0
turn off 499,499 through 500,500"""

data = get_data(day=6, year=2015).split('\n')
```

```python
DEBUG = False

if DEBUG:
    data = example.split('\n')
else:
    data = get_data(day=6, year=2015).split('\n')

lights = np.zeros([1000,1000]).astype('bool')
for instruction in data:
    print(instruction)
    instruction = instruction.replace('turn ','')
    instruction, start, _, end = instruction.split(' ')
    start = np.array(start.split(',')).astype('int')
    end = np.array(end.split(',')).astype('int') + 1
    if instruction == 'on':
        lights[start[0]:end[0],start[1]:end[1]] = True
    if instruction == 'off':
        lights[start[0]:end[0],start[1]:end[1]] = False
    if instruction == 'toggle':
        lights[start[0]:end[0],start[1]:end[1]] = ~lights[start[0]:end[0],start[1]:end[1]]
    

plt.imshow(lights)
```

```python
DEBUG = False

if DEBUG:
    data = example.split('\n')
else:
    data = get_data(day=6, year=2015).split('\n')

lights = np.zeros([1000,1000]).astype('int')
for instruction in data:
    print(instruction)
    instruction = instruction.replace('turn ','')
    instruction, start, _, end = instruction.split(' ')
    start = np.array(start.split(',')).astype('int')
    end = np.array(end.split(',')).astype('int') + 1
    if instruction == 'on':
        lights[start[0]:end[0],start[1]:end[1]] += 1
    if instruction == 'off':
        lights[start[0]:end[0],start[1]:end[1]] -= 1
    if instruction == 'toggle':
        lights[start[0]:end[0],start[1]:end[1]] += 2

    lights = np.fmax(lights,0)
    

plt.imshow(lights)
```

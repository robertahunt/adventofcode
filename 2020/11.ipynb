```python
from aocd import data, submit
from aocd.get import get_data
import numpy as np
import pandas as pd
import re
#Part 1
orig_data = get_data(day=11, year=2020).split('\n')
orig_data[:4]

for i in range(len(orig_data)):
    orig_data[i] = [x for x in orig_data[i]]
orig_data
```

```python
from copy import deepcopy

data = np.array(deepcopy(orig_data))
data_after_round = deepcopy(data)
rounds = 0
while (rounds < 1000):
    rounds += 1
    data_after_round = deepcopy(data)
    for i in range(len(data)):
        for j in range(len(data[0])):
            space = data[i,j]
            st_row = max(i-1,0)
            en_row = min(i+1,len(data))
            st_col = max(j-1,0)
            en_col = min(j+1,len(data[0]))

            if st_row == i:
                adj_row = 0
            elif en_row == i:
                adj_row = 1
            else:
                adj_row = 1
            if st_col == j:
                adj_col = 0
            elif en_col == j:
                adj_col = 1
            else:
                adj_col = 1



            adjacents = deepcopy(data[st_row:en_row+1,st_col:en_col+1])
            adjacents[adj_row,adj_col] = '.' #ignore current seat

            if space == '.': #floor
                continue
            elif space == 'L':
                if np.count_nonzero(adjacents == '#')  == 0:
                    data_after_round[i,j] = '#'
            elif space == '#':
                if np.count_nonzero(adjacents == '#') >= 4:
                    data_after_round[i,j] = 'L'

    if (data_after_round.flatten() == data.flatten()).all():
        break
    data = deepcopy(data_after_round)
data
```

```python
from copy import deepcopy
#part 2


data = np.array(deepcopy(orig_data))
data_after_round = deepcopy(data)
rounds = 0
while (rounds < 1000):
    rounds += 1
    data_after_round = deepcopy(data)
    for i in range(len(data)):
        for j in range(len(data[0])):
            space = data[i,j]
            st_row = max(i-1,0)
            en_row = min(i+1,len(data))
            st_col = max(j-1,0)
            en_col = min(j+1,len(data[0]))

            if st_row == i:
                adj_row = 0
            elif en_row == i:
                adj_row = 1
            else:
                adj_row = 1
            if st_col == j:
                adj_col = 0
            elif en_col == j:
                adj_col = 1
            else:
                adj_col = 1

            remove_cur = deepcopy(data)
            remove_cur[i,j] = '.' #ignore current seat
            adjacents = 0
            
            above, below, left, right, ul, ll, ur, lr = [np.array([])]*8
            
            above = data[:i,j][::-1]
            if i != len(data)-1:
                below = data[i+1:,j]
            
            left = data[i,:j][::-1]
            if j != len(data[0])-1:
                right = data[i,j+1:]
            
            ul = np.diag(data[:i,:j],j-i)[::-1]
            if (i != len(data)-1) & (j != len(data[0])-1):
                lr = np.diag(data[i+1:,j+1:])
            
            flip_i = i
            flip_j = len(data[0]) - 1 - j
            
            ur = np.diag(np.fliplr(data)[:i,:flip_j],flip_j-i)[::-1]
            if (i != len(data)-1) & (flip_j != len(data[0])-1):
                ll = np.diag(np.fliplr(data)[i+1:,flip_j+1:])
            
            adjacents =0
            for arr in [above, below, left, right, ul, ll, ur, lr]:
                for el in arr:
                    if el == '#':
                        adjacents += 1
                        break
                    elif el == 'L':
                        break
            
            if space == '.': #floor
                continue
            elif space == 'L':
                if adjacents == 0:
                    data_after_round[i,j] = '#'
            elif space == '#':
                if adjacents >= 5:
                    data_after_round[i,j] = 'L'

            #if (rounds == 3) & (i == 0) & (j == 3):
            #    asdf
    if (data_after_round.flatten() == data.flatten()).all():
        break
    data = deepcopy(data_after_round)
data
```

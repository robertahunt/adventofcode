```python
import os
import re
import aocd

data = aocd.get_data(day=18, year=2020).split('\n')
data
```

```python
# Part 1

# We want to first tackle the innermost brackets. 
# Let's find a regex that can find just the sets of brackets with no
# brackets inside
from tqdm import tqdm
from copy import deepcopy
def evaluate_result(e):
    #print('e_in',e)
    e = e.replace(')','').replace('(','')
    first_operation_pattern = '^[0-9]+[*+][0-9]+'
    operations = re.findall(first_operation_pattern, e)
    while len(operations):
        op = operations[0]
        #print('op',op)
        res = str(eval(op))
        #print('op_res',res)
        e = res + e[len(op):]
        operations = re.findall(first_operation_pattern, e)
    #print('e_out',e)
    return str(e)
    
#data = ['((2 + 4 * 9) * (6 + 9 * 8 + 6) + 6) + 2 + 4 * 2']
pattern = '(\([0-9*+]+?\))'
bracket_pattern = '\(+'
out = []
i = 0
for exp in tqdm(data):
    #print('Expression %s\n'%i)
    exp = deepcopy(exp)
    exp = exp.replace(' ','')
    while len(re.findall(bracket_pattern, exp)):
        #print('exp:',exp)
        mini_exp = re.findall(pattern, exp)[0]
        #print('mini_exp',mini_exp)
        result = evaluate_result(mini_exp)
        exp = exp.replace(mini_exp, result)
    #print('exp',exp)
    exp = evaluate_result(exp)
    out += [int(exp)]
    i += 1
    #if i == 10:
    #    break
import numpy as np
np.sum(out)
```

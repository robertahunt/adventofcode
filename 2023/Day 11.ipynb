```python
import os
import re
import aocd
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from copy import deepcopy
from aocd import submit

#Part 1
data = aocd.get_data(day=11, year=2023).split('\n')
```

```python
data = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....""".split('\n')
```

```python
expand_to = 1000000
data = np.array([[x for x in y] for y in data])
data[data == '.'] = 1
data[data == '#'] = 0
data = data.astype(int)
data
```

```python

for i in range(data.shape[0]):
    if 0 not in data[i]:
        data[i] = data[i]*expand_to
for i in range(data.shape[1]):
    if 0 not in data[:,i]:
        data[:,i] = data[:,i]*expand_to
data
```

```python
k = 1
pos = {} # dictionary of positions of each galaxy
for i in range(len(data)):
    for j in range(len(data.T)):
        if data[i,j] == 0:
            x_slice = data[:i,j]
            y_slice = data[i,:j]
            #print(x_slice, x_slice.sum(), len(x_slice[x_slice == 0]))
            
            x = x_slice.sum() + len(x_slice[x_slice == 0])
            y = y_slice.sum() + len(y_slice[y_slice == 0])
            pos[k] = [x,y]
            k += 1
pos
```

```python
# create an array of distances from each galaxy to each other
dists = np.zeros((k,k))
for i in range(1,k):
    for j in range(1,k):
        dists[i,j] = abs(pos[i][0] - pos[j][0]) + abs(pos[i][1] - pos[j][1])
dists = dists[1:,1:]

dists[np.triu_indices_from(dists, k=1)].sum()
```

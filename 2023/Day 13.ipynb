```python
import os
import re
import aocd
import string
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from copy import deepcopy
from aocd import submit
import pprint

#Part 1
data = aocd.get_data(day=13, year=2023).split('\n\n')
```

```python
data = """#.##..##.
..#.##.#.
##......#
##......#
..#.##.#.
..##..##.
#.#.##.#.

#...##..#
#....#..#
..##..###
#####.##.
#####.##.
..##..###
#....#..#""".split('\n\n')
```

```python
total = 0
for pattern in data:
    pattern = pattern.split('\n')
    pattern = np.array([[y for y in x] for x in pattern])
    print('\n')
    
    for l in pattern:
        print(l)
        
    # check horizontal
    for i in range(1,len(pattern)):
        start_row = max(0, i - (len(pattern) - i) )
        end_row = min(len(pattern), i + i)
        
        top = pattern[start_row:i]
        bottom = pattern[i:end_row]
        print('\nTop: \n',top)
        print('Bottom: \n',bottom)
        print((~(top == bottom[::-1])).sum())
        if (~(top == bottom[::-1])).sum() == 1:
            print(i)
            total += i*100
            break
        
    # check vertical
    for i in range(1,len(pattern[0])):
        start_col = max(0, i - (len(pattern[0]) - i) )
        end_col = min(len(pattern[0]), i + i)
        
        left = pattern[:,start_col:i]
        right = pattern[:,i:end_col]
        print('\nLeft: \n',left)
        print('Right: \n',right)
        if (~(left == right[:,::-1])).sum() == 1:
            print(i)
            total += i
            break
            
        
total
```

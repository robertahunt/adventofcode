```python
import os
import re
import aocd
import string
import numpy as np
import pandas as pd
from tqdm import tqdm
from copy import deepcopy
from aocd import submit


def is_lower(hand,current_lowest):
    if type_rank(hand) < type_rank(current_lowest):
        return True
    elif type_rank(hand) == type_rank(current_lowest):
        hand_dif, low_dif = first_different(hand,current_lowest)
        if order.index(hand_dif) > order.index(low_dif):
            #hand_dif is lower than low dif
            return True
    return False

def first_different(hand1, hand2):
    i = 0
    while hand1[i] == hand2[i]:
        i += 1
    return hand1[i], hand2[i]

def type_rank(hand):
    if five_of_a_kind(hand):
        return 7
    elif four_of_a_kind(hand):
        return 6
    elif full_house(hand):
        return 5
    elif three_of_a_kind(hand):
        return 4
    elif two_pair(hand):
        return 3
    elif one_pair(hand):
        return 2
    return 1

def n_of_a_kind(h, n):
    _, counts = np.unique([x for x in h], return_counts =True)
    if counts.max() == n:
        return True
    return False

def five_of_a_kind(h):
    if n_of_a_kind(h,n=5):
        return True
    return False

def four_of_a_kind(h):
    if n_of_a_kind(h,n=4):
        return True
    return False

def full_house(h):
    _, counts = np.unique([x for x in h], return_counts =True)
    if np.product(counts) == 6:
        return True
    

def three_of_a_kind(h):
    if n_of_a_kind(h,n=3):
        return True
    return False

def one_pair(h):
    if n_of_a_kind(h,n=2):
        return True
    return False

def two_pair(h):
    _, counts = np.unique([x for x in h], return_counts =True)
    if np.product(counts) == 4:
        return True
    return False


#Part 1
order = ['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']

data = aocd.get_data(day=7, year=2023).split('\n')

data = """32T3K 765
T55J5 684
KK677 28
KTJJT 220
QQQJA 483""".split('\n')

data = [x.split(' ') for x in data]

data = sorted(data,  key=lambda x: order.index(x[0][4]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][3]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][2]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][1]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][0]), reverse=True)
data = sorted(data,  key=lambda x: type_rank(x[0]), reverse=False)

#calculate the rank
for i in range(len(data)):
    data[i] += [type_rank(data[i][0]),i+1]
sum([int(x[1])*x[3] for x in data])
```

```python
# Part 2
def find_J_type_rank(h):
    best_type_rank = 1
    
    for new_card in order[:-1]:
        new_hand = h.replace('J',new_card)
        if type_rank(new_hand) >= best_type_rank:
            best_type_rank = type_rank(new_hand)
                    
    return best_type_rank 


data = aocd.get_data(day=7, year=2023).split('\n')

data = [x.split(' ') for x in data]

order = ['A', 'K', 'Q', 'T', '9', '8', '7', '6', '5', '4', '3', '2', 'J']

for i in range(len(data)):
    hand = data[i][0]
    if 'J' in hand:
        data[i] += [find_J_type_rank(hand)]
        
    else:
        data[i] += [type_rank(hand)]
data = sorted(data,  key=lambda x: order.index(x[0][4]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][3]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][2]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][1]), reverse=True)
data = sorted(data,  key=lambda x: order.index(x[0][0]), reverse=True)
data = sorted(data,  key=lambda x: x[2], reverse=False)

#calculate the rank
for i in range(len(data)):
    data[i] += [i+1]
sum([int(x[1])*x[3] for x in data])
```
